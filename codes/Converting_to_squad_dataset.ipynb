```python
from datasets import load_dataset, load_metric

datasets = load_dataset('json', 
          data_files={'train': '/content/drive/MyDrive/NLP_CA6/pquad_public/train_samples.json', 
          'test': '/content/drive/MyDrive/NLP_CA6/pquad_public/test_samples.json'}, field="data")
datasets

```

```python
dictionary= dict()
# Data to be written
dictionary['data'] = []

counter_id = [] 
data_ = datasets['test']
max_context_index = len(data_)

for index in range(max_context_index):

  qas = data_[index]['paragraphs'][0]['qas']
  for answer_index in range(len(qas)):
    if qas[answer_index]['is_impossible'] == False:
      

      temp_dict = dict()
      temp_dict['title'] = data_[index]['title']
      temp_dict['context'] = data_[index]['paragraphs'][0]['context']
      temp_asnwer_dict = dict()
      temp_asnwer_dict['answer_start'] = []
      temp_asnwer_dict['text'] = []
      if len(qas[answer_index]['answers']) == 1: 
        temp_asnwer_dict['answer_start'].append(qas[answer_index]['answers'][0]['answer_start'])
        temp_asnwer_dict['text'].append(qas[answer_index]['answers'][0]['text'])

      else : 
        for idx in range(len(qas[answer_index]['answers'])):
          temp_asnwer_dict['answer_start'].append(qas[answer_index]['answers'][idx]['answer_start'])
          temp_asnwer_dict['text'].append(qas[answer_index]['answers'][idx]['text'])
      temp_dict['answers']= temp_asnwer_dict
      temp_dict['id'] = qas[answer_index]['id']
      temp_dict['question'] = qas[answer_index]['question']
    
      dictionary['data'].append(temp_dict)







len(dictionary['data'])
# len(set(counter_id))
```

```python
import json
# Serializing json 
json_object = json.dumps(dictionary)
  
# Writing to sample.json
with open("/content/drive/MyDrive/NLP_CA6/new_jsonfile/pquad_public/test_samples.json", "w") as outfile:
    outfile.write(json_object)

# /content/drive/MyDrive/NLP_CA6/new_jsonfile/PersianQA/pqa_train.json
# /content/drive/MyDrive/NLP_CA6/new_jsonfile/PersianQA/pqa_test.json


# /content/drive/MyDrive/NLP_CA6/new_jsonfile/pquad_public/test_samples.json
# /content/drive/MyDrive/NLP_CA6/new_jsonfile/pquad_public/train_samples.json

# /content/drive/MyDrive/NLP_CA6/new_jsonfile/ParSQuAD/ParSQuAD-automatic-dev.json
# /content/drive/MyDrive/NLP_CA6/new_jsonfile/ParSQuAD/ParSQuAD-automatic-train.json
# /content/drive/MyDrive/NLP_CA6/new_jsonfile/ParSQuAD/ParSQuAD-manual-dev.json
# /content/drive/MyDrive/NLP_CA6/new_jsonfile/ParSQuAD/ParSQuAD-manual-train.json
```
